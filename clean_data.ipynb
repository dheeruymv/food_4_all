```python
def str_to_list(column_str):
  new_column_str = re.sub('[\[\]{}\'\"]', '', column_str)
  new_column_str = new_column_str.replace(", ", ",")
  return new_column_str.split(",")
```

```python
import re
column_names = ['ingredients', 'ingredients_raw_str', 'tags', 'search_terms']
for column_names in column_names:
  df[column_names] = df[column_names].apply(lambda x: str_to_list(x))
```

```python
df['steps'] = df['steps'].apply(lambda x: x.split("', '"))
df['steps'] = df['steps'].apply(lambda x: [re.sub('[\[\]{}\'\"]', '', step) for step in x])
```

```python
df['key_words'] = df['tags'] + df['search_terms']
```

```python
def add_filter_columns(df, filter_key_words_dict):
    for column_name, filter_key_words in filter_key_words_dict.items():
      for filter_key_word in filter_key_words:
        df.loc[df['key_words'].str.contains(filter_key_word, regex=False), column_name] = filter_key_word
    return df
```

```python
filter_key_words_dict = {
    'meals': ['breakfast', 'brunch', 'lunch', 'dinner', 'dessert', 'main-dish', 'picnic', 'smoothies', 'snack', 'pet-food'],
    'prepare_time': ['15-minutes-or-less', '30-minutes-or-less', '4-hours-or-less', '60-minutes-or-less', '1-day-or-more', '3-steps-or-less', '5-ingredients-or-less'],
    'ingredients_choice': ['low-calorie', 'low-carb', 'low-cholesterol', 'low-fat', 'low-protein', 'low-saturated-fat', 'low-sodium', 'low-sugar', 'lowfat', 'fat-free', 'gluten-free-appetizers', 'grain-free', 'no flour', 'no meat', 'no-shell-fish', 'non-alcoholic', 'sugar-free', 'sodium-free', 'toddler-friendly'],
    'food_types': ['beef','steak', 'pork', 'chicken', 'turkey', 'duck', 'egg', 'fish', 'shrimp', 'tuna', 'ham', 'lamb', 'lobster', 'tofu', 'octopus', 'oysters', 'vegan', 'vegetarian'],
}
```

```python
df = add_filter_columns(df, filter_key_words_dict)
```

```python
df.loc[df['key_words'].str.contains('ice-cream', regex=False), 'meals'] = 'dessert'
df.loc[df['key_words'].str.contains('cookie', regex=False), 'meals'] = 'dessert'
df.loc[df['key_words'].str.contains('lowfat', regex=False), 'ingredients'] = 'low-fat'
```
